```python
import pandas as pd
ratings=pd.read_csv('/content/Ratings.csv',encoding="latin-1",delimiter=",")
ratings
```

```python
books=pd.read_csv('/content/Books.csv',encoding="latin-1",delimiter=",")
books
```

```python
ratings_books=pd.merge(ratings,books,on="ISBN")
ratings_books.shape
```

```python
ratings_books
```

```python
books.shape
```

```python
ratings_books_sample=ratings_books.sample(frac=0.01,random_state=1)
ratings_books_sample.shape
```

```python
ratings_books_sample
```

```python
ratings_books_pivot=ratings_books_sample.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating').fillna(0)
ratings_books_pivot
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=7,n_jobs=-1)
model.fit(ratings_books_pivot)
```

```python
indices=model.kneighbors(ratings_books_pivot.loc[['10 Steps in Writing the Research Paper (10 Steps in Writing the Research Paper, 5th ed)']],10,return_distance=False)
indices
```

```python
for index,value in enumerate(ratings_books_pivot.index[indices.flatten()]):
    print((index+1),".",value)
```
